# Loading data

In this notebook, we will load some data from the Apple stock (aapl) from 1995 to 1998.

```python
import datetime
import pandas as pd
import pandas.io.data
from pandas import Series, DataFrame
from pandas.tools.plotting import scatter_matrix, autocorrelation_plot, radviz, andrews_curves
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
aapl = pd.io.data.get_data_yahoo('AAPL', 
                                 start=datetime.datetime(1995, 10, 1), 
                                 end=datetime.datetime(1998, 1, 1))
aapl.head()
```

We now plot some of these columns

```python
plt.plot(aapl['Open'])
plt.plot(aapl['Close'], color = 'r')
plt.show()
```

```python
aapl.plot(subplots=True, figsize=(6, 6)); 
```

```python
aapl['Volume'].hist()
```

```python
plt.boxplot(aapl['Open'])
```

Now, we create a scatter plot of all six columns. The (i,j) entry of this array corresponds to plotting the i-th column against the j-th column. We can see, with the exception of volume, there is a linear correlation between any pair of columns. On the diagonal, we can see the density plot for each column.

```python
scatter_matrix(aapl, alpha=0.2, figsize=(6, 6), diagonal='kde')
```

This is an autocorrelation plot for the openning value of the stock. The dashed band shows the 99% confidence interval, and the sold line the 95%. In the x-axis, we have values for different lags. 

```python
autocorrelation_plot(aapl['Open'])
```
